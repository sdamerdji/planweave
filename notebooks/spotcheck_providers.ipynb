```python
sample = df.sample(25, random_state=10).drop(columns=['County', 'Priority', 'Decision Maker', 'HCD Status'])
```

```python
sample['agendas_platform'] = None
```

```python
sample = sample.set_index('Jurisdiction')
```

```python
city_platforms = {
    "West Covina": "destinyhosted",
    "Sand City": "granicus",
    "Canyon Lake": "none?",
    "Millbrae": "laserfiche?",
    "Trinidad": "none?",
    "Belvedere": "granicus",
    "Biggs": "none",
    "Huron": "none",
    "Santa Paula": "granicus",
    "Gilroy": "granicus (and laserfiche)",
    "Pacific Grove": "destinyhosted",
    "San Joaquin": "none?",
    "Inglewood": "civicplus",
    "Sierra Madre": "civiclive",
    "Dixon": "granicus",
    "Fillmore": "granicus",
    "Cloverdale": "civicplus",
    "Concord": "none",
    "Del Rey Oaks": "civicplus",
    "Laguna Hills": "minutetraq (also granicus)",
    "Napa": "granicus",
    "Livermore": "granicus",
    "Montebello": "civicclerk",
    "Kerman": "civicplus"
}
```
